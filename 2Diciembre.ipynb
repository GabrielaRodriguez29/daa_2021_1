```
def fnRec(x):
  if x==0:
    print("stop")
  else :
    print(x)
    fnRec(x-1)
    
def main():
  print("inicio del programa")
  fnRec(5)
  print("Fin del programa")
main()
```

```
def printRev(x):
  if x>0:
    
    printRev(x-1)
    print(x)


printRev(3)
```

```
def fibonacci(n):
  if n==1 or n==0:
    return n
  else:
    return(fibonacci(n-1)+fibonacci(n-2))
fibonacci(8)
```
